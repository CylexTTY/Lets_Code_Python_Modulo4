## 2 - II -> Calcule se há diferença de média de número total de membros da família pelo sexo do chefe de família (Household Head Sex e Total Number of Family members) com 95% de confiança.

<h4>Definindo a hipótese nula como: Chefes de famílias do sexo MASCULINO possuem média de número de membros familiares IGUAL a chefes de famílias do sexo FEMININO.</h4>

<p>Então H0 = u0 - u1 = 0</p>

<p>H1 = u0 - u1 != 0</p>

```python
confianca = 0.95
significancia = 1- confianca
z_alpha_2 = norm.ppf(confianca + significancia / 2)
n_amostra
```

```python
['Household Head Sex', 'Total Number of Family members']
```

```python
len(df[df['Household Head Sex'] == 'Male']['Total Number of Family members'])
```

<h4>Aqui observa-se que o numero de mulheres e inferior ao numero amostral mínimo ideal</h4>

```python
len(df[df['Household Head Sex'] == 'Female']['Total Number of Family members'])
```

<h4>Para corrigir isso, poderia reduzir o tamanho da amostra ou entao, dar um replace=True para pegarmos resultados repetidos do restante que falta.</h4>

```python
df_mulheres = df[df['Household Head Sex'] == 'Female']['Total Number of Family members'].sample(n_amostra, random_state=101, replace=True)
df_homens = df[df['Household Head Sex'] == 'Male']['Total Number of Family members'].sample(n_amostra, random_state=101)
```

```python
media_H = df_homens.mean()
media_M = df_mulheres.mean()
```

```python
sigma_H = df_homens.std()
sigma_M = df_mulheres.std()
```

```python
nH = len(df_homens)
nM = len(df_mulheres)
```

```python
D0 = 0
```

<h4>Calculando Z</h4>

```python
numerador = (media_H - media_M) - D0
```

```python
denominador = np.sqrt((sigma_H**2 / nH) + (sigma_M**2 / nM))
```

```python
z = numerador / denominador
```

```python
z
```

<h4>Hipótese crítica:</h4>

```python
if (z <= - z_alpha_2) or (z >= z_alpha_2):
    print('Hipótese rejeitada.\n'
          'Há diferença de média de número total de membros da família pelo sexo do chefe de família')
else:
    print('Hipótese aceita.\n'
          'Não há diferença de média de número total de membros da família pelo sexo do chefe de família.')
```

```python
p_value = norm.sf(z)
p_value
```

```python
if p_value <= significancia:
     print('Hipótese rejeitada.\n'
          'Há diferença de média de número total de membros da família pelo sexo do chefe de família')
else:
    print('Hipótese aceita.\n'
          'Não há diferença de média de número total de membros da família pelo sexo do chefe de família.')
```

## 2 - III -> Verifique se há diferença de média total de renda da família (Total Household Income) de acordo com a fonte salarial (Main Source of Income).

<h4>Definindo a hipótese nula como: Renda salarial média IGUAL em todas fontes salariais.</h4>

<p>Então H0 = u0 - u1 = 0</p>

<p>H0 = u0 - u2 = 0</p>

<p>H0 = u1 - u2 = 0</p>

<p>H1 = u0 - u1 != 0</p>

<p>H1 = u0 - u2 != 0</p>

<p>H1 = u1 - u2 != 0</p>

<h4>Aqui a idéia é seguir como se fosse teste de duas amostras, porém, feito de 2 em 2, 3 vezes.</h4>

```python
df[['Total Household Income', 'Main Source of Income']].head()
```

```python
df['Main Source of Income'].unique()
```

```python
confianca = 0.95
significancia = 1- confianca
z_alpha_2 = norm.ppf(confianca + significancia / 2)
D0 = 0
```

```python
df_u0 = df[df['Main Source of Income'] == 'Wage/Salaries']['Total Household Income'].sample(n_amostra, random_state=101)
```

```python
df_u1 = df[df['Main Source of Income'] == 'Other sources of Income']['Total Household Income'].sample(n_amostra, random_state=101, replace=True)
```

```python
df_u2 = df[df['Main Source of Income'] == 'Enterpreneurial Activities']['Total Household Income'].sample(n_amostra, random_state=101, replace=True)
```

<h4>Dessa vez, irei utilizar a biblioteca Statsmodels.</h4>

```python
from statsmodels.stats.weightstats import DescrStatsW, CompareMeans
```

```python
teste_u0 = DescrStatsW(df_u0)
```

```python
teste_u1 = DescrStatsW(df_u1)
```

```python
teste_u2 = DescrStatsW(df_u2)
```

```python
teste1 = CompareMeans(teste_u0, teste_u1)
```

```python
z, p_value = teste1.ztest_ind(alternative='two-sided', value=D0)
```

```python
if (z <= - z_alpha_2) or (z >= z_alpha_2):
    print('Hipótese rejeitada.\n'
          'Há diferença de média total de renda da família de acordo com a fonte salarial.')
else:
    print('Hipótese aceita.\n'
          'Não há diferença de média total de renda da família de acordo com a fonte salarial.')
```

```python
if p_value <= significancia:
     print('Hipótese rejeitada.\n'
          'Há diferença de média total de renda da família de acordo com a fonte salarial.')
else:
    print('Hipótese aceita.\n'
          'Não há diferença de média total de renda da família de acordo com a fonte salarial..')
```

<h4>Basicamente não teria a necessidade de continuar o teste, visto que logo no primeiro experimento, a hipótese nula foi rejeitada. Porém, para fins de conclusão do projeto, irei fazer todas, porém de forma automatizada.</h4>

```python
lista_df = [(teste_u0, teste_u1), (teste_u0, teste_u2), (teste_u1, teste_u2)]
```

```python
cor = {'red': '\033[31m', 'green': '\033[32m', 'clean': '\033[m'}


for i, teste in enumerate(lista_df):
    teste_atual = CompareMeans(teste[0], teste[1])
    z, p_value = teste_atual.ztest_ind(alternative='two-sided', value=D0)

    print(f'\nTeste {i+1}')

    print('\nValor crítico "Z"')
    if (z <= - z_alpha_2) or (z >= z_alpha_2):
        print(f'{cor["red"]}Hipótese rejeitada{cor["clean"]}.\n'
          'Há diferença de média total de renda da família de acordo com a fonte salarial.')
    else:
        print(f'{cor["red"]}Hipótese aceita{cor["clean"]}.\n'
          'Não há diferença de média total de renda da família de acordo com a fonte salarial.')

    print('\nP_VALOR')
    if p_value <= significancia:
        print(f'{cor["red"]}Hipótese rejeitada{cor["clean"]}.\n'
          'Há diferença de média total de renda da família de acordo com a fonte salarial.')
    else:
        print(f'{cor["green"]}Hipótese aceita{cor["clean"]}.\n'
          'Não há diferença de média total de renda da família de acordo com a fonte salarial.')
```

<h4>Conclusão: Há diferença de média total de renda da família de acordo com a fonte salarial.</h4>

## 3 - I -> Usando Regressão Linear, crie um modelo de previsão de renda de forma a minimizar o erro. Considere 70% dos dados para treinar o modelo, e 30% para testá-lo. Por fim, utilize as métricas de avaliação do modelo que foram estudadas para validá-lo.

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.metrics import r2_score
```

```python
df_training = df[quantitativas].copy()
```

<h4>Removendo outliers</h4>

```python
df_training.corr()
```

```python
df_training_no_outliers = pd.DataFrame()
```

```python
for coluna in df_training.columns:
    q1 = df_training[coluna].quantile(0.25)
    q3 = df_training[coluna].quantile(0.75)

    IQR = q3 - q1

    inf = max(q1 - 1.5 * IQR, df_training[coluna].min())
    sup = min(q3 + 1.5 * IQR, df_training[coluna].max())

    df_training_no_outliers = df_training[(df_training[coluna] >= inf) & (df_training[coluna] <= sup)]
    
df_training_no_outliers.dropna(inplace=True)
```

```python
df_training_no_outliers.corr()
```

```python
X = df_training_no_outliers.drop('Total Household Income', axis='columns')
y = df_training_no_outliers['Total Household Income']
```

```python
x = sm.add_constant(X)
model = sm.OLS(y, x).fit()
```

<h4>Obtendo apenas as colunas com o P_Valor menor que 0.05 para assim obtermos as variáveis mais correlacionadas possíveis.</h4>

```python
correlacao_antes = 0
correlacao = [0]
```

```python
while correlacao_antes != len(correlacao):
    correlacao_antes = len(correlacao)

    correlacao = list(model.pvalues[model.pvalues < 0.05].index)

    correlacao.append('Total Household Income')

    if 'const' in correlacao:
        correlacao.remove('const')

    X = df_training_no_outliers[correlacao].drop('Total Household Income', axis='columns')
    y = df_training_no_outliers[correlacao]['Total Household Income']

    x = sm.add_constant(X)
    model = sm.OLS(y, x).fit()
```

```python
model.summary()
```
